```python
#change column Loan_ID to integer
data1['Loan_ID'] = data1['Loan_ID'].str.replace('LP', '').astype(int)
```

```python
data1.head()
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data1['Gender']=le.fit_transform(data1['Gender'])
data1['Married']=le.fit_transform(data1['Married'])
data1['Education']=le.fit_transform(data1['Education'])
data1['Self_Employed']=le.fit_transform(data1['Self_Employed'])
data1['Loan_Status']=le.fit_transform(data1['Loan_Status'])
data1['Property_Area']=le.fit_transform(data1['Property_Area'])

```

```python
data1.head()
```

```python
y=data1['Loan_Status']
x=data1.drop('Loan_Status',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
model=lin.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
y_pred = model.predict(x_test)
```

```python
loan_ids = x_test['Loan_ID']
```

```python
result_data1 = pd.DataFrame({'Loan_ID': loan_ids, 'Loan_Status': y_pred})
```

```python
result_data1.to_csv('loan_predictions.csv', index=False)
```
