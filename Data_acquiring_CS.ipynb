```python
q1=Fruit_prices['RetailPrice'].quantile(0.25)
q3=Fruit_prices['RetailPrice'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3-1.5*iqr
outlier=Fruit_prices[(Fruit_prices['RetailPrice']<lower_bound)] | Fruit_prices[(Fruit_prices['RetailPrice']>upper_bound)]
outlier
```

```python
mean_price = Fruit_prices['RetailPrice'].mean()
median_price = Fruit_prices['RetailPrice'].median()
price_75th_percentile = Fruit_prices['RetailPrice'].quantile(0.75)
price_75th_percentile = Fruit_prices['RetailPrice'].quantile(0.75)
print("Mean price:", mean_price)
print("Median price:", median_price)
print("75th percentile price:", price_75th_percentile)
```

```python
#1)Fruits with higher yields (more product per unit weight) tend to have higher retail prices.
#2)Fruits with larger cup equivalent sizes tend to have lower retail prices.
#3)Fruits with higher cup equivalent prices tend to have higher retail prices.
#These correlations provide insights into how various factors (yield, cup equivalent size, and cup equivalent price) are related to the retail prices of fruits in your dataset.
```
